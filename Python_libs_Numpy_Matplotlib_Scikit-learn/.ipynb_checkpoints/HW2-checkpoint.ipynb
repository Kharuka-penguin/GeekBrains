#### Задание 2

```python
t = np.linspace(0, 10, 51)
```

```python
f = np.cos(t)
```

```python
plt.plot(t, f, color='green')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

#### *Задание 3

```python
x =np.linspace(-3, 3, 51)
```

```python
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_title('График y1')
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title('График y2')

ax3.plot(x, y3)
ax3.set_title('График y3')

ax4.plot(x, y4)
ax4.set_title('График y4')
plt.show()
```

#### *Задание 4

```python
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
DATASET_PATH = './datasets/creditcard.csv'
df = pd.read_csv(DATASET_PATH, sep=',')
df.head()
```

```python
cards_fraud = df['Class'].value_counts()
cards_fraud
```

```python
cards_fraud.plot(kind='bar')
plt.show()
```

```python
cards_fraud.plot(kind='bar', logy=True)
plt.show()
```

```python
class_1 = df.loc[df['Class'] == 1, 'V1']
class_0 = df.loc[df['Class'] == 0, 'V1']
```

```python
plt.hist(class_1, bins=20, density=True, alpha=0.5, label='1')
plt.hist(class_0, bins=20, density=True, alpha=0.5, label='0')
plt.legend(loc='upper left')
plt.show()
```
