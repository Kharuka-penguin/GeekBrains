#### Задание 9

```python
from numpy import linalg as la

a = np.arange(0, 12)
A = a.reshape(4, 3)
At = A.T
B = np.matmul(A, At)
print(B,'\n\n', np.size(B))
b = la.det(B)
if b == 0:
    print('\nМатрица вырожденная, вычислить обратную матрицу нельзя')
else:
    print('\nОбратная матрица =', la.inv(B))
```
